```python
import pandas as pd
import numpy as np
from collections import Counter
import pprint as pp

df = pd.read_csv('data.csv', index_col=0)
df['Age10g'] = df['Age'] // 10


for i in range(1, 10):
    df_i = df[df['Age10g'] == i]   
    cnt = Counter(df_i['Department Name'])
    total = sum(cnt.values())
    keys, values = [], []
    for k in cnt: 
        keys.append(k)
        values.append(cnt[k] / total)
    df_i = pd.DataFrame({'class':keys, 'val':values})
    df_i.plot.bar(x='class', title='Grp %d' % i, subplots=True)
```
